```python
# 3: create churn_mlo_mdb.bronze_churn Delta table from the uploaded table
src_table = "mlops_catalog.default.wa_fn_use_c_telco_customer_churn"  # <- replace if needed
dest_table = "churn_mlo_mdb.bronze_churn"

spark.sql("CREATE DATABASE IF NOT EXISTS churn_mlo_mdb")

# drop & recreate so notebook is idempotent
spark.sql(f"DROP TABLE IF EXISTS {dest_table}")

spark.sql(f"""
  CREATE TABLE {dest_table}
  USING DELTA
  AS
  SELECT *, current_timestamp() AS ingestion_time
  FROM {src_table}
""")

print("Bronze table created:", dest_table)
display(spark.table(dest_table).limit(5))
```
